```python
import torch
from torch.distributions import MultivariateNormal, Normal
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="whitegrid")

torch.manual_seed(0)
```

# Bayesian Linear Regression - with known $\sigma$

## Model Definition
The Bayesian Linear Regression model assumes a linear relationship between input features and observed targets through:
$$y = Xw + \epsilon$$
where epsilon is noise, in this case assumed to be Gaussian: $\epsilon \sim N(0, \sigma^2 * I)$.
Here, $y = (y_1, ..., y_N)$, $X$ is the feature matrix of shape [N, D] (including a bias term), and $w$ is the weight vector of shape [D] (including bias).
Phrasing this in a Bayesian framework, we place a prior distribution over the weights w and define a likelihood for the observed data.
The likelihood is given by:
$$P(y | X, w) = N(y | Xw, \sigma^2 * I)$$
where:
- y: observed target vector
- X: feature matrix (with bias term)
- w: weight vector (including bias)
- $\sigma^2$: variance of the observation noise

And the prior over the weights is defined as:
$$P(w) = N(w | m_0, S_0)$$
where:
- $m_0$: prior mean vector
- $S_0$: prior covariance matrix

In this model we assume that $\sigma$ is known, otherwise we would need to place a prior over it and infer it as well in order to perform full Bayesian inference.

## Data Simulation

```python
def simulate_bayesian_linear_data(num_samples: int, num_features: int, noise_std: float,
                                  w_true: torch.Tensor=None) -> tuple[torch.Tensor, torch.Tensor, torch.Tensor]:
    """
    Generates data from a Bayesian Linear Regression model.

    This function follows the generative process:
    1. Define a prior for the weights (w).
    2. Sample the "true" weights (w_true) from this prior.
    3. Generate features (X), including a bias term.
    4. Generate observed labels (y) using y = Xw + epsilon,
       where epsilon is Gaussian noise.

    Args:
        num_samples (int): The number of data points (N).
        num_features (int): The number of explanatory features (D_features).
                            The total number of parameters will be D_features + 1 (for bias).
        noise_std (float): The standard deviation (sigma) of the observation noise.

    Returns:
        A tuple containing:
        - X (torch.Tensor): The [N, D_features + 1] feature matrix (with bias term).
        - y (torch.Tensor): The [N] observed target vector.
        - w_true (torch.Tensor): The [D_features + 1] "true" weight vector
                                 sampled from the prior.
    """

    # Total number of parameters (D) = num_features + 1 (for the bias/intercept)
    D_params = num_features + 1

    # 1. Define the prior distribution for the weights w.
    # We'll use a simple Gaussian prior centered at 0 with an identity covariance matrix.
    # P(w) = N(w | m_0, S_0)
    prior_mean = torch.zeros(D_params)
    prior_covariance = torch.eye(D_params)
    w_prior_dist = MultivariateNormal(prior_mean, prior_covariance)

    # 2. If w_true is not specified, sxample the "true" weights from the prior.
    w_true = w_prior_dist.sample() if w_true is None else w_true

    # 3. Generate the features X.
    # We'll generate the features from a standard Normal distribution.
    X_features = torch.randn(num_samples, num_features)
    #X_features = torch.poisson(torch.ones((num_samples, num_features)))

    # Add a column of ones for the bias/intercept term.
    X = torch.cat([torch.ones(num_samples, 1), X_features], dim=1)

    # 4. Generate the observed labels y.
    # Calculate the "true" mean response: y_true = X @ w_true
    y_true_mean = X @ w_true

    # Define the likelihood (observation noise) distribution
    # P(y | X, w) = N(y | Xw, sigma^2 * I)
    noise_dist = Normal(0.0, noise_std)

    # Sample the noise (epsilon)
    epsilon = noise_dist.sample(torch.Size([num_samples]))

    # Calculate the final observed labels
    y_observed = y_true_mean + epsilon

    return X, y_observed, w_true
```

### Simulation Parameters and Data Generation

```python
N_train = 10
N_test = 50
N_FEATURES = 1  # For a simple 2D plot (y = w_0 + w_1*x)
NOISE_STD_SIM = 0.5   # sigma
# --- Simulate Data ---
X, y, w_true = simulate_bayesian_linear_data(N_train, N_FEATURES, NOISE_STD_SIM)
X_test, y_test, _ = simulate_bayesian_linear_data(N_test, N_FEATURES, NOISE_STD_SIM, w_true)

X_full = torch.cat([X, X_test], dim=0)
y_full = torch.cat([y, y_test], dim=0)
```

## Bayesian Inference
Given the observed data (X, y) and the prior over weights, we can compute the posterior distribution over the weights w using Bayes' theorem:
$$P(w | X, y) \propto P(y | X, w) P(w)$$
For Bayesian Linear Regression with Gaussian priors and likelihoods, the posterior distribution is also Gaussian and can be computed in closed form:
$$P(w | X, y) = N(w | m_N, S_N)$$
where:
- $S_N = (S_0^{-1} + \frac{1}{\sigma^2} X^TX)^{-1}$
- $m_N = S_n (S_0^{-1} m_0 + \frac{1}{\sigma^2} X^Ty)$

### Infer Posterior from data

Below is a function that computes the posterior mean and covariance given the data and prior parameters as defined above.

```python
def infer_posterior(
    X: torch.Tensor,
    y: torch.Tensor,
    prior_mean: torch.Tensor,
    prior_cov: torch.Tensor,
    noise_std: float
) -> tuple[torch.Tensor, torch.Tensor]:
    """
    Infers the posterior distribution of w given data (X, y) and a prior.

    Implements the closed-form solution for Bayesian Linear Regression
    with a Gaussian prior and Gaussian likelihood.

    P(w | X, y) = N(w | m_N, S_N)

    Args:
        X (torch.Tensor): The [N, D] feature matrix (with bias term).
        y (torch.Tensor): The [N] observed target vector.
        prior_mean (torch.Tensor): The prior mean (m_0).
        prior_cov (torch.Tensor): The prior covariance (S_0).
        noise_std (float): The standard deviation (sigma) of the observation noise.

    Returns:
        A tuple containing:
        - posterior_mean (torch.Tensor): The posterior mean (m_N).
        - posterior_cov (torch.Tensor): The posterior covariance (S_N).
    """

    noise_var = noise_std**2

    # Calculate the posterior covariance (S_N)
    # S_N^-1 = S_0^-1 + (1/sigma^2) * X.T @ X
    prior_precision = torch.inverse(prior_cov)
    data_precision = (1 / noise_var) * X.T @ X
    posterior_precision = prior_precision + data_precision
    posterior_cov = torch.inverse(posterior_precision)

    # Calculate the posterior mean (m_N)
    # m_N = S_N @ (S_0^-1 @ m_0 + (1/sigma^2) * X.T @ y)
    weighted_prior_mean = prior_precision @ prior_mean
    weighted_data = (1 / noise_var) * X.T @ y
    posterior_mean = posterior_cov @ (weighted_prior_mean + weighted_data)

    return posterior_mean, posterior_cov
```

#### Inference parameters

```python
# --- 2. Infer Posterior ---
NOISE_STD = NOISE_STD_SIM   # Set to the same noise used in data simulation
D_params = N_FEATURES + 1
PRIOR_MEAN = torch.zeros(D_params)
PRIOR_COV = torch.eye(D_params) * 2.0  # Use a slightly wider prior
posterior_mean, posterior_cov = infer_posterior(
        X, y, PRIOR_MEAN, PRIOR_COV, NOISE_STD
    )
```

#### MLE Estimate for comparison

```python
# Compute MLE estimate for comparison
XTX_inv = torch.inverse(X.T @ X)
w_mle = XTX_inv @ X.T @ y
print("MLE Weights:\n", w_mle)
```

### Visualize posterior over weights

```python
# Visualizes the posterior N(w | m_N, S_N) on the weight space
# Compare with MLE estimate and true weights
# Only for 1D feature case (2D weight space including bias)
if N_FEATURES == 1:
    print("\nPosterior Mean:\n", posterior_mean)
    print("\nPosterior Covariance:\n", posterior_cov)

    # Create a grid of weight values
    grid_offset0 = 3 * torch.sqrt(posterior_cov[0, 0])
    grid_offset1 = 3 * torch.sqrt(posterior_cov[1, 1])
    w0 = torch.linspace(posterior_mean[0] - grid_offset0, posterior_mean[0] + grid_offset0, 100)
    w1 = torch.linspace(posterior_mean[1] - grid_offset1, posterior_mean[1] + grid_offset1, 100)
    W0, W1 = torch.meshgrid(w0, w1, indexing='ij')
    pos = torch.stack([W0.ravel(), W1.ravel()], dim=-1)
    # Compute the posterior density at each point in the grid
    posterior_dist = MultivariateNormal(posterior_mean, posterior_cov)
    Z = torch.exp(posterior_dist.log_prob(pos)).reshape(W0.shape)
    # Plot the posterior density
    plt.figure(figsize=(10, 6))
    plt.contourf(W0.numpy(), W1.numpy(), Z.numpy(), levels=50, cmap='viridis')
    plt.colorbar(label='Posterior Density')
    # True weights, MLE estimate, and posterior mean
    plt.scatter(w_true[0].item(), w_true[1].item(), color='red', marker='x', s=100, label='True Weights')
    plt.scatter(w_mle[0].item(), w_mle[1].item(), color='yellow', marker='^', s=100, label='MLE Estimate')
    plt.scatter(posterior_mean[0].item(), posterior_mean[1].item(), color='white', marker='o', s=100, label='Posterior Mean')
    plt.title('Posterior Distribution over Weights', fontsize=16)
    plt.xlabel('Weight w_0 (Bias)', fontsize=12)
    plt.ylabel('Weight w_1', fontsize=12)
    plt.legend(fontsize=10)
```

### Posterior weights comment

The posterior mean plotted above be seen as a regularized version of the MLE estimate, where the prior pulls the estimate towards zero (or the prior mean).
It gives us a formal way to incorporate prior beliefs about the weights, which may be derived by domain knowledge or previous data.

### Visualize Posterior Inference
We can visualize the posterior inference by plotting the observed data, the true underlying line, and the posterior mean line. Additionally, we can illustrate the uncertainty in the parameter estimates by plotting confidence intervals and samples from the posterior distribution.

```python
# --- 3. Plotting (for 1D feature case) ---
if N_FEATURES == 1:
    plt.figure(figsize=(12, 8))

    # Plot the observed data points
    plt.scatter(X[:, 1].numpy(), y.numpy(), alpha=0.6, label="Observed Data (y)")

    # Create a line of points to plot on
    x_line = torch.linspace(X[:, 1].min() - 0.5, X[:, 1].max() + 0.5, 100)
    x_line_with_bias = torch.cat([torch.ones(100, 1), x_line.unsqueeze(1)], dim=1)

    # Plot the "true" underlying line
    y_true_line = x_line_with_bias @ w_true
    plt.plot(x_line.numpy(), y_true_line.numpy(), color='red', linestyle='--',
             linewidth=2.5, label=f"True Line")

    # Plot the posterior mean line
    y_post_mean_line = x_line_with_bias @ posterior_mean
    plt.plot(x_line.numpy(), y_post_mean_line.numpy(), color='blue',
             linewidth=2.5, label="Posterior Mean Line")

    # Plot the MLE line for comparison
    y_mle_line = x_line_with_bias @ w_mle
    plt.plot(x_line.numpy(), y_mle_line.numpy(), color='yellow',
             linewidth=2.5, label="MLE Line")

    # --- Plot Parameter Uncertainty (Confidence Interval for the line) ---
    parameter_variance = torch.diag(x_line_with_bias @ posterior_cov @ x_line_with_bias.T)
    parameter_std = torch.sqrt(parameter_variance)

    plt.fill_between(
        x_line.numpy(),
        (y_post_mean_line - 2 * parameter_std).numpy(),
        (y_post_mean_line + 2 * parameter_std).numpy(),
        color='blue', # Shading matches the posterior mean line
        alpha=0.15,
        label="95% Confidence Interval (Parameter Uncertainty)"
    )

    # --- Plot Samples from Posterior (Parameter Uncertainty) ---
    posterior_dist = MultivariateNormal(posterior_mean, posterior_cov)
    num_posterior_samples_to_plot = 5
    for i in range(num_posterior_samples_to_plot):
        w_sample = posterior_dist.sample()
        y_sample_line = x_line_with_bias @ w_sample
        plt.plot(x_line.numpy(), y_sample_line.numpy(), color='green',
                 alpha=0.4, linewidth=1,
                 label="Posterior Samples" if i == 0 else None) # Only label once

    plt.title("Bayesian Linear Regression: Posterior Inference", fontsize=16)
    plt.xlabel("Feature (x)", fontsize=12)
    plt.ylabel("Label (y)", fontsize=12)
    plt.legend(fontsize=10)
    plt.grid(True)
    plt.ylim(y.min() - 1, y.max() + 1)

    # Save or show the plot
    print("Plot saved to 'blr_inference_plot.png'")
    # plt.show()
```

## Predictive Posterior
The predictive distribution for a new input $X_{new}$ is given by:
$$P(y_{new} | X_{new}, X, y) = \int_W P(y_{new}, w | X_{new}, X, y) dw = \int P(y_{new} | X_{new}, w) P(w | X, y) dw$$
For Bayesian Linear Regression with known $\sigma^2$, this integral can be computed in closed form, resulting in:
$$P(y_{new} | X_{new}, X, y) = N(y_{new} | m^{pred}, S^{pred})$$
where:
- $m^{pred} = X_{new} m_N$
- $S^{pred} = \sigma^2 * I + X_{new} S_N X_{new}^T$

This captures both the uncertainty in the weight estimates (epistemic uncertainty) and the inherent observation noise from $\sigma$ (aleatoric uncertainty).

We can also use this expression to show how predictions with the MLE weights are done by setting $P(w | X, y) = \delta(w - w^{MLE})$:
$$P_{MLE}(y_{new} | X_{new}, X, y) = \int P(y_{new} | X_{new}, w) \delta(w - w^{MLE}) dw = P(y_{new} | X_{new}, w^{MLE}) = N(y_{new} | X_{new} w^{MLE}, \sigma^2 * I)$$

```python
def predictive_posterior(
    X_new: torch.Tensor,
    posterior_mean: torch.Tensor,
    posterior_cov: torch.Tensor,
    noise_std: float
) -> tuple[torch.Tensor, torch.Tensor]:
    """
    Computes the predictive posterior distribution for new inputs.

    P(y_new | X_new, X, y) = N(y_new | X_new @ m_N, sigma^2 * I + X_new @ S_N @ X_new.T)

    Args:
        X_new (torch.Tensor): The [M, D] new feature matrix (with bias term).
        posterior_mean (torch.Tensor): The posterior mean (m_N).
        posterior_cov (torch.Tensor): The posterior covariance (S_N).
        noise_std (float): The standard deviation (sigma) of the observation noise.

    Returns:
        A tuple containing:
        - predictive_mean (torch.Tensor): The predictive mean for new inputs.
        - predictive_cov (torch.Tensor): The predictive covariance for new inputs.
    """

    noise_var = noise_std**2

    # Predictive mean
    predictive_mean = X_new @ posterior_mean

    # Predictive covariance
    predictive_cov = noise_var * torch.eye(X_new.shape[0]) + X_new @ posterior_cov @ X_new.T

    return predictive_mean, predictive_cov
```

```python
# --- 4. Predictive Posterior ---
# Calculate m^pred and S^pred for test inputs
predictive_mean, predictive_cov = predictive_posterior(X_test, posterior_mean, posterior_cov, NOISE_STD)
```

### Visualize Predictive Posterior

```python
if N_FEATURES == 1:
    print("\nGenerating Predictive Posterior plot...")
    sns.set_theme(style="whitegrid")
    plt.figure(figsize=(12, 8))

    # Plot the training data
    plt.scatter(X[:, 1].numpy(), y.numpy(), alpha=0.6,
                label="Training Data", color='blue')
    # Plot the test data
    plt.scatter(X_test[:, 1].numpy(), y_test.numpy(), alpha=0.6,
                label="Test Data", color='green', marker='^')

    # Create a smooth line of points to plot on
    # Use the min/max of all data for a good range
    all_x = torch.cat([X[:, 1], X_test[:, 1]])
    x_line = torch.linspace(all_x.min() - 0.5, all_x.max() + 0.5, 100)
    x_line_with_bias = torch.cat([torch.ones(100, 1), x_line.unsqueeze(1)], dim=1)

    # --- Calculate Predictive Distribution for the smooth line ---
    # This is the correct way to get the data for the plot
    line_pred_mean, line_pred_cov = predictive_posterior(
        x_line_with_bias, posterior_mean, posterior_cov, NOISE_STD
    )

    # We only need the variance (diagonal) for the shading
    predictive_variance = torch.diag(line_pred_cov)
    predictive_std = torch.sqrt(predictive_variance)

    # Plot the predictive mean line (size 100)
    plt.plot(x_line.numpy(), line_pred_mean.numpy(), color='orange',
             linewidth=2.5, label="Predictive Mean Line")

    # Plot the predictive uncertainty shading (size 100)
    plt.fill_between(
        x_line.numpy(), # <-- Size 100
        (line_pred_mean - 2 * predictive_std).numpy(), # <-- Size 100
        (line_pred_mean + 2 * predictive_std).numpy(), # <-- Size 100
        color='orange',
        alpha=0.2,
        label="95% Predictive Interval"
    )

    plt.title("Bayesian Linear Regression: Posterior Predictive", fontsize=16)
    plt.xlabel("Feature (x)", fontsize=12)
    plt.ylabel("Label (y)", fontsize=12)
    plt.legend(fontsize=10)
    plt.grid(True)

    # Set ylim based on all data
    all_y = torch.cat([y, y_test])
    plt.ylim(all_y.min() - 1, all_y.max() + 1)

    # Save or show the plot
    print("Plot saved to 'blr_predictive_plot.png'")
    # plt.show()
```
